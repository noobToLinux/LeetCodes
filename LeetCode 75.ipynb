## Merge Strings Alternately

You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. If a string is longer than the other, append the additional letters onto the end of the merged string.

Return the merged string.

Example:

* w1: 'abc', w2: 'pqr' -> 'apbqcr'
* w1: 'ab', w2: 'pqrs' -> 'apbqrs'
* w1: 'abcd', w2: 'pq' -> 'apbqcd'

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        newword = ''
        for i in range(max(len(word1),len(word2))):
            try:
                l1 = word1[i]
            except IndexError:
                l1 = ''
            
            try:
                l2 = word2[i]
            except IndexError:
                l2 = ''
            
            newword = newword + l1 + l2

        return newword
```

## Greatest Common Divisor of Strings

For two strings s and t, we say "t divides s" if and only if s = t + ... + t (i.e., t is concatenated with itself one or more times).

Given two strings str1 and str2, return the largest string x such that x divides both str1 and str2.

Example:

* s1: 'ABCABC', s2: 'ABC' -> 'ABC'
* s2: 'ABABAB', s2: 'ABAB' -> 'AB'
* s3: 'Leet', s2: 'Code' -> ''

```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        lp1 = self.listofcd(str1)
        lp2 = self.listofcd(str2)
        gCommon = self.common(lp1,lp2)
        if len(gCommon)>0:
            return gCommon[-1]
        else:
            return ''
    
    def gcd(self,str1:str) -> str:
        for i in range(len(str1)-1,0,-1):
            pattern = str1[:i]
            if str1 == pattern*int((len(str1)/len(pattern))):
                return pattern
        return str1
    
    def mcd(self,str1:str) -> str:
        for i in range(1,len(str1)):
            pattern = str1[:i]
            if str1 == pattern*int((len(str1)/len(pattern))):
                return pattern
        return str1
    
    def listofcd(self,str1:str) -> str:
        LOCD = list()
        for i in range(len(str1),0,-1):
            pattern = str1[:i]
            if str1 == pattern*int((len(str1)/len(pattern))):
                LOCD.append(pattern)
        return sorted(LOCD,key=len)
    
    def common(self,l1:list,l2:list) -> list:
        return [e for e in l1 if e in l2]
```

## Kids with the greatest number of candies

There are n kids with candies. You are given an integer array candies, where each candies\[i\] represents the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.

Return a boolean array result of length n, where result\[i\] is true if, after giving the ith kid all the extraCandies, they will have the greatest number of candies among all the kids, or false otherwise.

Note that multiple kids can have the greatest number of candies.

Example:

* candies = [2,3,5,1,3], extraCandies = 3 -> [true,true,true,false,true]
* candies = [4,2,1,1,2], extraCandies = 1 -> [true,false,false,false,false]
* candies = [12,1,12], extraCandies = 10 -> [true,false,true]

Other description:

Must be True if after giving the extra candies, the kid will have more than the others (without giving any to the others), return false otherwise

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        mv = max(candies)
        return [nc+extraCandies>=mv for nc in candies]
```

## Reverse vowels of a string

Given a string s, reverse only all the vowels in the string and return it.

The vowels are 'a', 'e', 'i', 'o', and 'u', and they can appear in both lower and upper cases, more than once.

Example:

* s='hello' -> 'holle'
* s='leetcode' -> 'leotcede'

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        ns = ''
        vowels = list('aeiouAEIOU')
        rvowels = list()
        for i in range(1,len(s)+1):
            l = s[-i]
            if l in vowels:
                rvowels.append(l)
        I = 0
        for i in range(len(s)):
            l = s[i]
            if l in vowels:
                ns = ns + rvowels[I]
                I=I+1
            else:
                ns = ns + l
        
        return ns
```

## Reverse Words in a String

Given an input string s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.

Return a string of the words in reverse order concatenated by a single space.

Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should only have a single space separating the words. Do not include any extra spaces.

Example:

* s = 'the sy is blue' -> 'blue is sky the'
* s = ' hello world ' -> 'world hello'
* s = 'a good  example' -> 'example good a'

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        #assuming words are separated by spaces
        lwords = list(filter(lambda n: n != '', s.split(' ')))
        ns = ''
        for i,w in enumerate(lwords[::-1]):
            if i==0:
                ns = ns + w
            else:
                ns = ns + ' ' + w
        return ns
```

## Product of Arry except self

Given an integer array nums, return an array answer such that answer\[i\] is equal to the product of all the elements of nums except nums\[i\].

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in O(n) time and without using the division operation.

Example:

* nums = \[1,2,3,4\] -> \[24,12,8,6\]
* nums = \[-1,1,0,-3,3\] -> \[0,0,9,0,0\]

**sudesh_pawar solution explanation**

### Intuition
Like any other problem, my first intuition was let's brute force the solution by having two pointers going from i-1 to 1 and i+1 to len(nums) and multiplying each element and then appending this result to the output array. However, on secondary thought (and mainly due to solving questions such as Prefix Sum and Postfix Sum before), I could observe that we can use similar concept like Prefix and Postfix.

### Approach
We use the fact that prefix_product of arr\[i\] is arr\[0\] * arr\[1\] * .. * arr\[i-1\] and postfix_product of arr\[i\] is arr\[i+1\] * arr\[i+2\] * .. * arr\[n-1\].

So basically, we have to calculate prefix_product * postfix_product\[i\] for each element.

Most solutions implementing the concept of Prefix and Postfix would suggest 2 traversals, however I felt that we could one-up that and come up with a single for-loop solution.

1. Initialize a Solution Array of same size as input array with value.
2. Store Prefix and Postfix Product so far in variables.
3. Traverse the input array.
4. Before updating the values for each i, multiply current solution array value at i with the value of prefix i.e. multiply with prefix product of the previous i-1 elements.
5. Similarly, calculate the postfix product value for n-i-1 where n is length of input array at each iteration.
6. As in Step 4, before calculating the postfix for i'th value , multiply the solution_array\[n-i-1\] with the postfix product value i.e. products of input\[i+1\] to input\[n-1\].

Please do like the solution if you understood it. Helps boosting visibility :-P

### Complexity
* Time complexity: O(n) Single Pass \[Should technically be 100\% faster than other solutions but Leetcode showing only 85\%. Don't know why\]
* Space complexity: Technically O(1) as we are not supposed to count the output array.


```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        length=len(nums)
        sol=[1]*length
        pre = 1
        post = 1
        for i in range(length):
            sol[i] *= pre
            pre = pre*nums[i]
            sol[length-i-1] *= post
            post = post*nums[length-i-1]
        return(sol)
```

## String compression

Given an array of characters chars, compress it using the following algorithm:

Begin with an empty string s. For each group of consecutive repeating characters in chars:

* If the group's length is 1, append the character to s.
* Otherwise, append the character followed by the group's length.

The compressed string s should not be returned separately, but instead, be stored in the input character array chars. Note that group lengths that are 10 or longer will be split into multiple characters in chars.

After you are done modifying the input array, return the new length of the array.

You must write an algorithm that uses only constant extra space.

Example:

* chars = \["a","a","b","b","c","c","c"\] -> Return 6, and the first 6 characters of the input array should be: \["a","2","b","2","c","3"\]
    * The groups are "aa", "bb", and "ccc". This compresses to "a2b2c3".
* chars = \["a"\] -> Return 1, and the first character of the input array should be: \["a"\]
    * The only group is "a", which remains uncompressed since it's a single character.
* chars = \["a","b","b","b","b","b","b","b","b","b","b","b","b"\] -> Return 4, and the first 4 characters of the input array should be: \["a","b","1","2"\].
    * The groups are "a" and "bbbbbbbbbbbb". This compresses to "ab12".

```python
class Solution:
    def compress(self, chars: List[str]) -> int:
        nel = len(chars)
        count = 1
        char = chars.pop(0)
        s = ''        
        for _ in range(nel-1):
            c = chars.pop(0)
            if c!= char:
                if count>1:
                    s = s + char + str(count)
                else:
                    s = s + char
                char = c
                count = 1
            else:
                count = count+1
        
        if count>1:
            s = s + char + str(count)
        else:
            s = s + char
        
        chars.extend(list(s))

        return None
```

## Is subsequence

Given two strings s and t, return true if s is a subsequence of t, or false otherwise.

A subsequence of a string is a new string that is formed from the original string by deleting some (can be none) of the characters without disturbing the relative positions of the remaining characters. (i.e., "ace" is a subsequence of "abcde" while "aec" is not).

Examples:

* s = "abc", t = "ahbgdc" -> True
* s = "axc", t = "ahbgdc" -> False

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        j=0
        ls = len(s)
        lt = len(t)
        if ls == 0:
            return True
        if ls>lt:
            return False
        v2 = s[0]
        for i in range(lt):
            v1 = t[i]
            if v1 == v2:
                if j == len(s)-1:
                    return True
                j=j+1
                v2 = s[j]
        return False
```

## Container with most water

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height\[i\]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

Example:
* height = \[1,8,6,2,5,4,8,3,7\] -> 49
    * The above vertical lines are represented by array \[1,8,6,2,5,4,8,3,7\]. In this case, the max area of water (blue section) the container can contain is 49.
* height = \[1,1\] -> 1

### Explanation

#### Intuition:
The two-pointer technique starts with the widest container and moves the pointers inward based on the comparison of heights.
Increasing the width of the container can only lead to a larger area if the height of the new boundary is greater. By moving the pointers towards the center, we explore containers with the potential for greater areas.

#### Explanation:
1. Initialize the variables:

    * left to represent the left pointer, starting at the beginning of the container (index 0).
    * right to represent the right pointer, starting at the end of the container (index height.size() - 1).
    * maxArea to keep track of the maximum area found, initially set to 0.
2. Enter a loop using the condition left < right, which means the pointers have not crossed each other yet.

3. Calculate the current area:

    * Use the min function to find the minimum height between the left and right pointers.
    * Multiply the minimum height by the width, which is the difference between the indices of the pointers: (right - left).
    * Store this value in the currentArea variable.
4. Update the maximum area:

    * Use the max function to compare the currentArea with the maxArea.
    * If the currentArea is greater than the maxArea, update maxArea with the currentArea.
5. Move the pointers inward: (Explained in detail below)

    * Check if the height at the left pointer is smaller than the height at the right pointer.
    * If so, increment the left pointer, moving it towards the center of the container.
    * Otherwise, decrement the right pointer, also moving it towards the center.
6. Repeat steps 3 to 5 until the pointers meet (left >= right), indicating that all possible containers have been explored.

7. Return the maxArea, which represents the maximum area encountered among all the containers.

#### Update the maximum area:
The purpose of this condition is to determine which pointer to move inward, either the left pointer (i) or the right pointer (j), based on the comparison of heights at their respective positions.

Imagine you have two containers represented by the heights at the left and right pointers. The condition checks which container has a smaller height and moves the pointer corresponding to that container.

1. If height\[i\] > height\[j\]:

    * This means that the height of the left container is greater than the height of the right container.
    * Moving the right pointer (j) would not increase the potential area because the height of the right container is the limiting factor.
    * So, to explore containers with the possibility of greater areas, we need to move the right pointer inward by decrementing j.
2. If height\[i\] <= height\[j\]:

    * This means that the height of the left container is less than or equal to the height of the right container.
    * Moving the left pointer (i) would not increase the potential area because the height of the left container is the limiting factor.
    * So, to explore containers with the possibility of greater areas, we need to move the left pointer inward by incrementing i.
By making these pointer movements, we ensure that we are always exploring containers with the potential for larger areas. The approach is based on the observation that increasing the width of the container can only lead to a larger area if the height of the new boundary is greater.

By following this condition and moving the pointers accordingly, the algorithm explores all possible containers and finds the one with the maximum area.

```python
class Solution:
    def maxArea2(self, height: List[int]) -> int:
        Area = lambda index1,index2: min(height[index1],height[index2])*abs(index2-index1)
        PArea = lambda index1,index2: max(height[index1],height[index2])*abs(index2-index1)
        def dAI(ind1,ind2,first : bool):
            AO = Area(ind1,ind2)
            if first:
                A2 = Area(ind1+1,ind2)
            if not first:
                A2 = Area(ind1,ind2-1)
            return A2-AO
        
        def dPAI(ind1,ind2,first : bool):
            AO = PArea(ind1,ind2)
            if first:
                A2 = PArea(ind1+1,ind2)
            if not first:
                A2 = PArea(ind1,ind2-1)
            return A2-AO

        i=0
        j=len(height)-1
        Ar = 0
        while i<j:
            PreAr = Ar
            Ar = max(Area(i,j),Ar)
            deltaA1 = dAI(i,j,True)
            deltaA2 = dAI(i,j,False)
            deltaPA1 = dPAI(i,j,True)
            deltaPA2 = dPAI(i,j,False)
            #print(deltaA1,deltaA2,deltaPA1,deltaPA2,':',i,j,':',Ar,PreAr)
            L = [(0,deltaA1),(1,deltaA2),(2,deltaPA1),(3,deltaPA2)]
            best = max(L,key=lambda x: x[1])
            if best[1]<0:
                if best[0] in [0,1]:
                    j=j+1
                else:
                    i=i-1
            elif best[0] in [0,1]:
                if deltaA1>deltaA2:
                    i=i+1
                else:
                    j=j-1
            else:
                if deltaPA2>deltaPA1: #Watch 2 and 1 are switched here
                    i=i+1
                else:
                    j=j-1

        return Ar
    
    def maxArea(self, height: List[int]) -> int:
        #rahulvarma5297 solution
        left = 0
        right = len(height) - 1
        maxArea = 0

        while left < right:
            currentArea = min(height[left], height[right]) * (right - left)
            maxArea = max(maxArea, currentArea)

            if height[left] < height[right]:
                left += 1
            else:
                right -= 1

        return maxArea
```

## Max number of k-sum pairs

You are given an integer array nums and an integer k.

In one operation, you can pick two numbers from the array whose sum equals k and remove them from the array.

Return the maximum number of operations you can perform on the array.

Example:

* nums = \[1,2,3,4\], k = 5 -> 2
    * Starting with nums = \[1,2,3,4\]:
        * Remove numbers 1 and 4, then nums = \[2,3\]
        * Remove numbers 2 and 3, then nums = \[\]
    * There are no more pairs that sum up to 5, hence a total of 2 operations.
* nums = \[3,1,3,4,3\], k = 6 -> 1
    * Starting with nums = \[3,1,3,4,3\]:
        * Remove the first two 3's, then nums = \[1,4,3\]
    * There are no more pairs that sum up to 6, hence a total of 1 operation.

### Explanation

#### Intuition
The intuition behind this solution is to use a two-pointer approach on a sorted array. Sorting the array allows us to efficiently find pairs of numbers that sum up to the target value k. The two pointers (left and right) are initialized at the beginning and end of the sorted array, and we move them towards each other based on the sum of the numbers at those positions.

#### Approach
Sort the array: We begin by sorting the input array nums. Sorting is essential for the two-pointer approach. using the nums.sort()

Iterate with two pointers: Use a while loop to iterate until left pointer is less than right pointer.
- If the sum of nums\[left\] and nums\[right\] is equal to the target k, we have found a pair, so we increment the operation count and move both pointers towards the center
- If the sum is less than k, we need a larger sum, so we move the left pointer to the right.
- If the sum is greater than k, we need a smaller sum, so we move the right pointer to the left.

Return the count: After the while loop completes, return the operation count, which represents the maximum number of pairs that can be formed to achieve the target sum.

#### Complexity
* Time complexity:
    * The time complexity is dominated by the sorting step, which is typically O(n log n) for the average case. The two-pointer traversal of the sorted array is O(n), so the overall time complexity is O(n log n)

* Space complexity:
    * The space complexity is O(1) because the algorithm uses a constant amount of extra space for the pointers and variables, regardless of the input size.

```python
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        gto = lambda x:x>=0
        Si = lambda x: 1 if gto(x) else -1

        nl = len(nums)
        i=0
        j=nl-1
        counter=0
        nums.sort()
        while i<j:
            v1 = nums[i]
            v2 = nums[j]
            s=v1+v2
            if v1>k:
                i=i+1
            elif v2>k:
                j=j-1
            else:
                if s==k:
                    i=i+1
                    j=j-1
                    counter+=1
                elif s<k:
                    i=i+1
                else:
                    j=j-1
        return counter
```

# Sliding window

## Maximum average subarray I

You are given an integer array nums consisting of n elements, and an integer k.

Find a contiguous subarray whose length is equal to k that has the maximum average value and return this value. Any answer with a calculation error less than 10-5 will be accepted.

Example:
* nums = \[1,12,-5,-6,50,3\], k = 4 -> 12.75000
    * Maximum average is (12 - 5 - 6 + 50) / 4 = 51 / 4 = 12.75
* nums = \[5\], k=1 -> 5.00000

### Explanation

#### Approach

This problem uses sliding window concept.
You can solve this problem using two loops in O(n2)O(n^2)O(n 
2
 ) time complexity. However, with sliding window approach, you can easily resolve this in O(n) time.

Follow the images below to look at how this works:


1. Our given input: We initialize calculating the sum of the first k elements:
    * Since we have to find the max average, we can use just the sum of out comparison.
2. Next iteration:
    * Once we slide the window, the new sum will always be the current sum + the new element - the element that left.
    * And we store the max of the previous and the current sum.
3. We get the final result by divinding by k, the size of the window.


```python
class Solution:
    def findMaxAverage2(self, nums: List[int], k: int) -> float:
        nl = len(nums)
        if k>=nl:
            return sum(nums)/nl
        i = 0
        fi = lambda x: x+k-1
        mean=float('-inf')
        while fi(i)<nl:
            mean = max(mean,sum(nums[i:fi(i)+1])/k)
            i=i+1
        return mean
    
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        #younus_baig
        # Initialize currSum and maxSum to the sum of the initial k elements
        currSum = maxSum = sum(nums[:k])

        # Start the loop from the kth element 
        # Iterate until you reach the end
        for i in range(k, len(nums)):

            # Subtract the left element of the window
            # Add the right element of the window
            currSum += nums[i] - nums[i - k]
            
            # Update the max
            maxSum = max(maxSum, currSum)

        # Since the problem requires average, we return the average
        return maxSum / k
```

Return the size of the longest non-empty subarray containing only 1's in the resulting array. Return 0 if there is no such subarray.

Examples:

    * After deleting the number in position 2, \[1,1,1\] contains 3 numbers with value of 1's.
* nums = \[0,1,1,1,0,1,1,0,1\] -> 5
    * After deleting the number in position 4, \[0,1,1,1,1,1,0,1\] longest subarray with value of 1's is \[1,1,1,1,1\].
* nums = \[1,1,1\] -> 2

```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        intervalsL = list()
        counter = 1
        for i,n in enumerate(nums):
            if i == 0:
                continue
            pren = nums[i-1]
            if n != pren:
                intervalsL.append((pren,counter))
                counter = 1
            else:
                counter += 1
        
        intervalsL.append((n,counter))
        
        maxsize = 0
        sI = len(intervalsL)-1
        if sI+1 == 1:
            if intervalsL[0][0]==1:
                return intervalsL[0][1]-1
            else:
                return 0
        elif sI+1 == 2:
            if intervalsL[0][0] == 1:
                return intervalsL[0][1]
            else:
                return intervalsL[1][1]

        for i,(v,s) in enumerate(intervalsL):
            if i == 0 or i == sI:
                continue
            prev,pres = intervalsL[i-1]
            posv,poss = intervalsL[i+1]

            c = v == 0 and s>1

            if c:
                maxsize = max(maxsize,pres,poss)
            elif v==0:
                maxsize = max(maxsize,pres+poss)
            else:
                maxsize = max(maxsize,s)
        

        return maxsize
        
    def longestSubarray2(self, nums: List[int]) -> int:
        #Better memory efficient
        ans = 0
        left = 0
        zeros = 0

        for right in range(len(nums)):
            if nums[right] == 0:
                zeros += 1
            while zeros > 1 and left <= right:
                if nums[left] == 0:
                    zeros -= 1
                left += 1
            ans = max(ans, right - left)

        return ans
    
    def longestSubarray3(self, nums: List[int]) -> int:
        #Faster solution
        i = 0
        j = 0
        maxLen = 0
        curr = 0
        zero = 0
        while i < len(nums):
            if nums[i]==0:
                zero += 1
                while zero > 1:
                    if nums[j]==0:
                        zero -= 1
                    else:
                        curr -= 1
                    j += 1
            else:
                curr += 1
                maxLen = max(maxLen,curr)
            i += 1 
        if maxLen == len(nums): 
            maxLen -= 1 
        return maxLen
```

## Find pivot index

Given an array of integers nums, calculate the pivot index of this array.

The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the index's right.

If the index is on the left edge of the array, then the left sum is 0 because there are no elements to the left. This also applies to the right edge of the array.

Return the leftmost pivot index. If no such index exists, return -1.

Examples:

* Input: nums = \[1,7,3,6,5,6\] -> 3
    * The pivot index is 3. Left sum = nums\[0\] + nums\[1\] + nums\[2\] = 1 + 7 + 3 = 11 Right sum = nums\[4\] + nums\[5\] = 5 + 6 = 11
* Input: nums = \[1,2,3\] -> -1
    * There is no index that satisfies the conditions in the problem statement.
* Input: nums = \[2,1,-1\] -> 0
    * The pivot index is 0. Left sum = 0 (no elements to the left of index 0) Right sum = nums\[1\] + nums\[2\] = 1 + -1 = 0

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        nl = len(nums)
        lsum = 0
        rsum = sum(nums)
        for i,n in enumerate(nums):
            rsum -= n
            if lsum == rsum:
                return i
            lsum += n
        return -1
```

Given two 0-indexed integer arrays nums1 and nums2, return a list answer of size 2 where:
* answer\[0\] is a list of all distinct integers in nums1 which are not present in nums2.
* answer\[1\] is a list of all distinct integers in nums2 which are not present in nums1.
Note that the integers in the lists may be returned in any order.

Examples:

    * For nums1, nums1\[1\] = 2 is present at index 0 of nums2, whereas nums1\[0\] = 1 and nums1\[2\] = 3 are not present in nums2. Therefore, answer\[0\] = \[1,3\]. For nums2, nums2\[0\] = 2 is present at index 1 of nums1, whereas nums2\[1\] = 4 and nums2\[2\] = 6 are not present in nums2. Therefore, answer\[1\] = \[4,6\].
* nums1 = \[1,2,3,3\], nums2 = \[1,1,2,2\] -> \[\[3\],\[\]\]
    * For nums1, nums1\[2\] and nums1\[3\] are not present in nums2. Since nums1\[2\] == nums1\[3\], their value is only included once and answer\[0\] = \[3\]. Every integer in nums2 is present in nums1. Therefore, answer\[1\] = \[\].


```python
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        #Beats 95% in speed but only 14% in memory
        nums1.sort()
        nums2.sort()

        nums_1 = [nums1[0]]
        nums_2 = [nums2[0]]
        #Remove duplicates
        for i,n in enumerate(nums1):
            if i == 0:
                continue
            pren = nums1[i-1]
            if n != pren:
                nums_1.append(n)

        for i,n in enumerate(nums2):
            if i == 0:
                continue
            pren = nums2[i-1]
            if n != pren:
                nums_2.append(n)
        
        L1 = list()
        L2 = list()
        i,j = 0,0
        EndI = False
        EndJ = False
        while i<len(nums_1) or j<len(nums_2):
            try:
                n1 = nums_1[i]
            except IndexError:
                n1 = nums_1[-1]
                EndI = True
            try:
                n2 = nums_2[j]
            except IndexError:
                n2 = nums_2[-1]
                EndJ = True
            
            if n1<n2:
                if EndI:
                    L1.append(n2)
                    j+=1
                else:
                    L2.append(n1)
                    i+=1
            elif n2<n1:
                if EndJ:
                    L2.append(n1)
                    i+=1
                else:
                    L1.append(n2)
                    j+=1
            else:
                i+=1
                j+=1

        return [L2,L1]

    
    def findDifference2(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        #Most common answer, a bit slower
        set1 = set(nums1)
        set2 = set(nums2)

        answer = set1 - set2, set2 - set1
        return answer
    
    def findDifference3(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        #Better memory efficient
        res1 = [ ]
        res2 = [ ]
        common = []
        for i in nums1:
            if i in nums2:
                common.append(i)
            else:
                if i not in res1:
                    res1.append(i)
        for j in nums2:
            if j not in common and j not in res2:
                res2.append(j)
        return [res1, res2]
```

## Unique number of ocurrences

Given an array of integers arr, return true if the number of occurrences of each value in the array is unique or false otherwise.

Examples:

* arr = \[1,2,2,1,1,3\] -> True
    * The value 1 has 3 occurrences, 2 has 2 and 3 has 1. No two values have the same number of occurrences.
* arr = \[1,2\] -> False
* arr = \[-3,0,1,-3,1,1,1,-3,10,0\] -> True

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        counterdict = dict()
        for n in arr:
            try:
                counterdict[n]
            except KeyError:
                counterdict[n] = 0
            counterdict[n] += 1
        
        Vs = counterdict.values()
        return len(Vs) == len(set(Vs))
        
```

## Determine if two strings are close

Two strings are considered close if you can attain one from the other using the following operations:

* Operation 1: Swap any two existing characters.
    * For example, abcde -> aecdb
* Operation 2: Transform every occurrence of one existing character into another existing character, and do the same with the other character.
    * For example, aacabb -> bbcbaa (all a's turn into b's, and all b's turn into a's)

You can use the operations on either string as many times as necessary.

Given two strings, word1 and word2, return true if word1 and word2 are close, and false otherwise.

* word1 = "abc", word2 = "bca" -> True
    * You can attain word2 from word1 in 2 operations. Apply Operation 1: "a**bc**" -> "a**cb**" Apply Operation 1: "**a**c**b**" -> "**b**c**a**"
* word1 = "a", word2 = "aa" -> False
    * It is impossible to attain word2 from word1, or vice versa, in any number of operations.
* word1 = "cabbba", word2 = "abbccc" -> True
    * You can attain word2 from word1 in 3 operations. Apply Operation 1: "ca**b**bb**a**" -> "ca**a**bb**b**" Apply Operation 2: "**c**aa**bbb**" -> "**b**aa**ccc**" Apply Operation 2: "**baa**ccc" -> "**abb**ccc"

```python
class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:
        dictunique1 = dict()
        dictunique2 = dict()

        for l in word1:
            try:
                dictunique1[l]
            except KeyError:
                dictunique1[l] = 0
            dictunique1[l]+=1
        
        for l in word2:
            try:
                dictunique2[l]
            except KeyError:
                dictunique2[l] = 0
            dictunique2[l]+=1
        
        if sorted(list(dictunique1.keys())) != sorted(list(dictunique2.keys())):
            return False
        if sorted(list(dictunique1.values())) != sorted(list(dictunique2.values())):
            return False
        return True
```

## Equal row and column pairs

Given a 0-indexed n x n integer matrix grid, return the number of pairs (ri, cj) such that row ri and column cj are equal.

A row and column pair is considered equal if they contain the same elements in the same order (i.e., an equal array).

* grid = \[\[3,2,1\],\[1,7,6\],\[2,7,7\]\] -> 1
    * There is 1 equal row and column pair:
        * (Row 2, Column 1): \[2,7,7\]
* grid = \[\[3,1,2,2\],\[1,4,4,5\],\[2,4,2,2\],\[2,4,2,2\]\] -> 3
    * There are 3 equal row and column pairs:
        * (Row 0, Column 0): \[3,1,2,2\]
        * (Row 2, Column 2): \[2,4,2,2\]
        * (Row 3, Column 2): \[2,4,2,2\]


```python
class Solution:
    def equalPairs(self, grid: List[List[int]]) -> int:
        counter = 0
        n = len(grid)
        cols = list()
        for c in range(n):
            cols.append([grid[f][c] for f in range(n)])

        for row in grid:
            for col in cols:
                if row == col:
                    counter += 1
        return counter
```

Note:

The input will be generated such that the operation is always possible.
It can be shown that the resulting string will always be unique.
 

Examples:

* s = "leet**cod*e" -> 'lecoe'
    * Performing the removals from left to right:
        - The closest character to the 1st star is 't' in "leet**cod*e". s becomes "lee*cod*e".
        - The closest character to the 2nd star is 'e' in "lee*cod*e". s becomes "lecod*e".
        - The closest character to the 3rd star is 'd' in "lecod*e". s becomes "lecoe".
        - There are no more stars, so we return "lecoe".
* s = "erase*****" -> ''
    * The entire string is removed, so we return an empty string.

```python
class Solution:
    def removeStars(self, s: str) -> str:
        Intervals=list()
        counter = 0
        prel = 'a' if s[0] != '*' else '*'
        Init = 0
        ########### Step 1 ###################
        for i,ly in enumerate(s):
            if ly != '*':
                l='a'
            else:
                l='*'
            if l != prel:
                End = i-1
                if prel != '*':
                    Intervals.append(('a',counter,Init,End))
                else:
                    Intervals.append(('*',counter,Init,End))
                counter = 0
                Init = i
            counter += 1
            prel = l
        #We add the last element
        End = i
        if prel != '*':
            Intervals.append(('a',counter,Init,End))
        else:
            Intervals.append(('*',counter,Init,End))
        ########### Step 2 ###################
        if Intervals[0][0] == '*':
            Intervals = Intervals[1:]
        FL = list()
        for i,I in enumerate(Intervals):
            if I[0] != '*':
                continue
            preI = Intervals[i-1]
            nI = (preI[0],preI[1]-I[1],preI[2])
            FL.append(nI)
        #we add the last one
        if I[0] != '*':
            nI = (I[0],I[1],I[2])
            FL.append(nI)
        
        ########### Step 3 ###################
        ne = len(FL)
        ns = ''
        ssum = 0
        for i,fI in enumerate(FL):
            ssum += fI[1]
            if i == ne-1:
                continue
            if fI[1] == 0:
                continue
            elif fI[1]<0:
                ns = ns[:fI[1]]
            else:
                ns += s[fI[2]:fI[2]+fI[1]]
            if ssum == 0:
                ns = ''
        if fI[1]>0:
            ns += s[fI[2]:fI[2]+fI[1]]
        elif fI[1]<0:
            ns = ns[:fI[1]]
        return ns

    def removeStars2(self, s: str) -> str:
        #Faster and simpler
        word = []
        for char in s:
            if char == '*':
                word.pop()
            else:
                word.append(char)

        return ''.join(word)
```

# Asteroid Collision

We are given an array asteroids of integers representing asteroids in a row.

For each asteroid, the absolute value represents its size, and the sign represents its direction (positive meaning right, negative meaning left). Each asteroid moves at the same speed.

Find out the state of the asteroids after all collisions. If two asteroids meet, the smaller one will explode. If both are the same size, both will explode. Two asteroids moving in the same direction will never meet.

Examples:

* asteroids = \[5,10,-5\] -> \[5,10\]
    * The 10 and -5 collide resulting in 10. The 5 and 10 never collide.
* asteroids = \[8,-8\] -> \[\]
    * The 8 and -8 collide exploding each other.
* asteroids = \[10,2,-5\] -> \[10\]
    * The 2 and -5 collide resulting in -5. The 10 and -5 collide resulting in 10.

```python
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        
        NL = [asteroids[0]]
        i=1
        while i<len(asteroids):
            if len(NL) == 0: #This is for the moment it empty itself
                NL.append(asteroids[i])
                i+=1
                continue
            LA = NL[-1]
            A = asteroids[i]
            ##################################
            if self.CollisionCheck(LA,A):
                EX = self.ExplosionCheck(LA,A)
                ##############################
                if EX == 0:
                    NL.pop() # Eliminamos el último elemento
                elif EX == 1:
                    pass
                elif EX == 2:
                    NL.pop()
                    i-=1 #This is to maintain the i value
                ##############################
            else:
                NL.append(A)
            ##################################
            i+=1
        return NL

    
    def CollisionCheck(self,A1,A2):
        if self.Direction(A1)<0:
            return False
        elif self.Direction(A1) == self.Direction(A2):
            return False
        else:
            return True
    
    def ExplosionCheck(self,A1,A2):
        if self.Size(A1)>self.Size(A2):
            return 1
        if self.Size(A2)>self.Size(A1):
            return 2
        else:
            return 0
    
    def Size(self,A):
        return abs(A)
    
    def Direction(self,A):
        if A>0:
            return 1
        elif A<0:
            return -1
        else:
            return None
    
    def asteroidCollision2(self, asteroids: List[int]) -> List[int]:
        #Faster solution
        stack = []
        for asteroid in asteroids:
            stack.append(asteroid)
            
            while (len(stack) >= 2 and stack[-1] < 0 and stack[-2] > 0):
                collidingAsteroid1 = stack.pop(-1)
                collidingAsteroid2 = stack.pop(-1)
                if abs(collidingAsteroid1) > abs(collidingAsteroid2):
                    stack.append(collidingAsteroid1)
                elif abs(collidingAsteroid1) < abs(collidingAsteroid2):
                    stack.append(collidingAsteroid2)
        return stack
```

## Decode string

Given an encoded string, return its decoded string.

The encoding rule is: k\[encoded_string\], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2\[4\].

The test cases are generated so that the length of the output will never exceed 10^5.

Examples:

* s = "3\[a\]2\[bc\]" -> "aaabcbc"
* s = "3\[a2\[c\]\]" -> "accaccacc"
* s = "2\[abc\]3\[cd\]ef" -> "abcabccdcdcdef

```python
class Solution:
    def decodeString(self, s: str) -> str:
        # Better than 97.69% of py3 users in speed, 7.14% in memory
        counter = 0
        Numbers = list('0123456789')
        Brackets = list('[]')
        fs = ''
        NTimes = ''
        for i,l in enumerate(s):
            if l in Numbers and counter == 0:
                NTimes += l
            elif l not in Numbers and l not in Brackets and counter == 0:
                fs += l
            elif l == '[':
                counter += 1
                if counter == 1:
                    Ini = i
                    if NTimes == '':
                        NTimes = 1
            elif l==']':
                counter -= 1
                if counter == 0:
                    End = i
                    fs += int(NTimes)*self.decodeString(s[Ini+1:End])
                    NTimes = ''
        return fs
```

```python
class RecentCounter:

    def __init__(self):
        self.Requests = [[]]
        self.LastIndex = 1
        

    def ping(self, t: int) -> int:
        LastIndexUpdated = False
        Ini = t-3000
        End = t
        Interval = [Ini,End]
        counter = 1
        for i in range(self.LastIndex,len(self.Requests)):
            I = self.Requests[i]
            if Ini > I[1]:
                continue
            elif End < I[0]:
                continue
            elif Ini <= I[1]:
                if not LastIndexUpdated:
                    self.LastIndex = i
                    LastIndexUpdated = True
                counter += 1
        
        self.Requests.append(Interval)
        return counter

    ## Much faster
    def __init__2(self):
        self.q = deque()
        
    def ping2(self, t: int) -> int:
        self.q.append(t)
        
        while t - self.q[0] > 3000:
            self.q.popleft()
            
        return len(self.q)
```

## Dota2 senate

In the world of Dota2, there are two parties: the Radiant and the Dire.

The Dota2 senate consists of senators coming from two parties. Now the Senate wants to decide on a change in the Dota2 game. The voting for this change is a round-based procedure. In each round, each senator can exercise one of the two rights:

Ban one senator's right: A senator can make another senator lose all his rights in this and all the following rounds.
Announce the victory: If this senator found the senators who still have rights to vote are all from the same party, he can announce the victory and decide on the change in the game.
Given a string senate representing each senator's party belonging. The character 'R' and 'D' represent the Radiant party and the Dire party. Then if there are n senators, the size of the given string will be n.

The round-based procedure starts from the first senator to the last senator in the given order. This procedure will last until the end of voting. All the senators who have lost their rights will be skipped during the procedure.

Suppose every senator is smart enough and will play the best strategy for his own party. Predict which party will finally announce the victory and change the Dota2 game. The output should be "Radiant" or "Dire".

Examples:

* senate = "RD" -> "Radiant"
    * The first senator comes from Radiant and he can just ban the next senator's right in round 1. 
    * And the second senator can't exercise any rights anymore since his right has been banned. 
    * And in round 2, the first senator can just announce the victory since he is the only guy in the senate who can vote.
* senate = "RDD" -> "Dire"
    * The first senator comes from Radiant and he can just ban the next senator's right in round 1. 
    * And the second senator can't exercise any rights anymore since his right has been banned. 
    * And the third senator comes from Dire and he can ban the first senator's right in round 1. 
    * And in round 2, the third senator can just announce the victory since he is the only guy in the senate who can vote.

```python
class Solution:
    def predictPartyVictory(self, senate: str) -> str:
        TotalR = 0
        TotalD = 0
        NRBanned = 0
        NDBanned = 0
        NR = 0 # We use this variables to store the left survivors
        ND = 0
        newsenate = ''
        preS = 'A'
        #-# LOOP 1 #-#
        for l in senate:
            #print(TotalR,TotalD,':',l,':',NR,ND,':',NRBanned,NDBanned,':',preS)
            if l == 'R':
                if NRBanned>0:
                    NRBanned -= 1
                else:
                    NDBanned += 1
                    TotalR += 1
                    NR += 1
                #if we changed from party we add the survivors and reset the counter
                if preS == 'D': # this second condition ensures that they are added in case all in such row have been removed
                    newsenate += ND*'D'
                    ND = 0
            elif l == 'D':
                if NDBanned>0:
                    NDBanned -= 1
                else:
                    NRBanned += 1
                    TotalD += 1
                    ND += 1
                if preS == 'R':
                    newsenate += NR*'R'
                    NR = 0
            preS = l
        else:
            if preS == 'R':
                newsenate += NR*'R'
            elif preS == 'D':
                newsenate += ND*'D'

        if TotalR <= 0:
            return 'Dire'
        elif TotalD <= 0:
            return 'Radiant'
        
        #-# LOOP 2 #-#
        #Now we move in reverse removing the remaining bans from the start
        newsenate2 = ''
        for l in newsenate:
            if l == 'R':
                if NRBanned>0:
                    NRBanned -= 1
                    TotalR -= 1
                else:
                    newsenate2 += l
            if l == 'D':
                if NDBanned>0:
                    NDBanned -= 1
                    TotalD -= 1
                else:
                    newsenate2 += l
        
        if TotalR == 0:
            return 'Dire'
        elif TotalD == 0:
            return 'Radiant'
        else:
            return self.predictPartyVictory(newsenate2)
```

## Delete the middle node of a linked list

You are given the head of a linked list. Delete the middle node, and return the head of the modified linked list.

The middle node of a linked list of size n is the ⌊n / 2⌋th node from the start using 0-based indexing, where ⌊x⌋ denotes the largest integer less than or equal to x.

For n = 1, 2, 3, 4, and 5, the middle nodes are 0, 1, 1, 2, and 2, respectively.

Examples:
* head = \[1,3,4,7,1,2,6\] -> \[1,3,4,1,2,6\]
    * The above figure represents the given linked list. The indices of the nodes are written below.
    * Since n = 7, node 3 with value 7 is the middle node, which is marked in red.
    * We return the new list after removing this node.
* head = \[1,2,3,4\] -> \[1,2,4\]
    * The above figure represents the given linked list.
    * For n = 4, node 2 with value 3 is the middle node, which is marked in red.
* head = \[2,1\] -> \[2\]
    * The above figure represents the given linked list.
    * For n = 2, node 1 with value 1 is the middle node, which is marked in red.
    * Node 0 with value 2 is the only node remaining after removing node 1.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Beats 95.84% in speed and 96.51% in memory of Python users.
        middlenode = 0
        counter = 0
        v = head
        # We get the middle node
        while v != None:
            counter += 1
            v=v.next
        middlenode = int(round(counter/2-0.1))
        # we remove the middle node
        index = 0
        v = head
        if middlenode == 0:
            return None
        while v != None:
            if index == middlenode-1:
                v.next = v.next.next
                break
            v = v.next
            index += 1

        return head
```

## Odd even linked list

Given the head of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return the reordered list.

The first node is considered odd, and the second node is even, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problem in O(1) extra space complexity and O(n) time complexity.

Examples:

* head = \[1,2,3,4,5\] -> \[1,3,5,2,4\]
* head = \[2,1,3,5,6,4,7\] -> \[2,3,6,7,1,5,4\]

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        odd = head
        if odd == None:
            return None
        firsteven = head.next
        if firsteven == None:
            return odd
        while odd.next != None:
            even = odd.next
            odd.next = odd.next.next
            if even.next == None:
                break
            even.next = even.next.next
            #
            odd = odd.next
        odd.next = firsteven
        return head
```

## Reverse linked list

Given the head of a singly linked list, reverse the list, and return the reversed list.

Examples:
* head = \[1,2\] -> \[2,1\]
* head = \[\] -> \[\]

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        n = head
        if head == None:
            return None
        prev = None
        while n != None:
            temp = n.next
            n.next = prev
            prev = n
            n = temp
        return prev
```

    * Nodes 0 and 1 are the twins of nodes 3 and 2, respectively. All have twin sum = 6.
    * There are no other nodes with twins in the linked list.
    * Thus, the maximum twin sum of the linked list is 6.
* head = [4,2,2,3] -> 7
    * The nodes with twins present in this linked list are:
        * Node 0 is the twin of node 3 having a twin sum of 4 + 3 = 7.
        * Node 1 is the twin of node 2 having a twin sum of 2 + 2 = 4.
    * Thus, the maximum twin sum of the linked list is max(7, 4) = 7.
* head = [1,100000] -> 100001
    * There is only one node with a twin in the linked list having twin sum of 1 + 100000 = 100001.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    # Beats 96.26 and 83.51 in speed and memory
    def pairSum(self, head: Optional[ListNode]) -> int:
        n = head
        rev,middlenode = self.ReverseAndMiddleNode(head)
        Index = 0
        maxsum = 0
        while Index < middlenode:
            maxsum = max(maxsum,rev[Index]+n.val)
            n = n.next
            Index += 1
        return maxsum
    
    def ReverseAndMiddleNode(self,head):
        n = head
        prev = None
        counter = 0
        TL = list()
        while n != None:
            counter += 1
            TL.append(n.val)
            n = n.next
        middlenode = int(round(counter/2-0.1))
        return TL[::-1],middlenode
```

## Maximum depth of binary tree

Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.


Examples:

* root = \[3,9,20,null,null,15,7\] -> 3
* root = \[1,null,2\] -> 2

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        self.maxdepth = 0
        if root is None:
            return 0
        self.IncreaseDepth(root,1)
        return self.maxdepth

    def IncreaseDepth(self,Node,Depth):
        if Node.left is not None:
            self.IncreaseDepth(Node.left,Depth+1)
        if Node.right is not None:
            self.IncreaseDepth(Node.right,Depth+1)
        self.maxdepth = max(self.maxdepth,Depth)
        return Depth
    
    def maxDepth2(self, root: Optional[TreeNode]) -> int:
        #Simpler and faster solution
        if not root:
            return 0
        return 1+max(self.maxDepth(root.right), self.maxDepth(root.left))
```

## Leaf-Similar Trees

Consider all the leaves of a binary tree, from left to right order, the values of those leaves form a leaf value sequence.

For example, in the given tree above, the leaf value sequence is (6, 7, 4, 9, 8).

Two binary trees are considered leaf-similar if their leaf value sequence is the same.

Return true if and only if the two given trees with head nodes root1 and root2 are leaf-similar.

Examples:

* root1 = \[3,5,1,6,2,9,8,null,null,7,4\], root2 = \[3,5,1,6,7,4,2,null,null,null,null,null,null,9,8\] -> True
* root1 = \[1,2,3\], root2 = \[1,3,2\] -> False

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def leafSimilar(self, root1: Optional[TreeNode], root2: Optional[TreeNode]) -> bool:
        #Beats 99.61% at runtime but 6.07 at memory
        L1 = self.LeafValue(root1)
        L2 = self.LeafValue(root2)
        
        if len(L1) != len(L2):
            return False
        for i,V1 in enumerate(L1):
            V2 = L2[i]
            if V1 != V2:
                return False
        return True
    
    def LeafValue(self,Node):
        Leaves = list()
        if Node is None:
            Leaves.extend([])
        elif Node.left is None and Node.right is None:
            Leaves.extend([Node.val])
        else:
            Leaves.extend(self.LeafValue(Node.left))
            Leaves.extend(self.LeafValue(Node.right))
        return Leaves
```

## Count good nodes in binary tree

Given a binary tree root, a node X in the tree is named good if in the path from root to X there are no nodes with a value greater than X.

Return the number of good nodes in the binary tree.

Examples:

* root = \[3,1,4,3,null,1,5\] -> 4
    * Nodes in blue are good.
    * Root Node (3) is always a good node.
    * Node 4 -> (3,4) is the maximum value in the path starting from the root.
    * Node 5 -> (3,4,5) is the maximum value in the path
    * Node 3 -> (3,1,3) is the maximum value in the path.
* root = \[3,3,null,4,2\] -> 3
    * Node 2 -> (3, 3, 2) is not good, because "3" is higher than it
* root = \[1\] -> 1
    * Root is considered as good.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def goodNodes(self, root: TreeNode) -> int:
        #Beats 94.62% in memory, 5.15% at runtime
        self.counter = 0
        self.Check(root,root.val)
        return self.counter
    
    def Check(self,Node,PrevMaxValue):
        if Node is None:
            return 0
        if Node.val>=PrevMaxValue:
            self.counter += 1
            self.Check(Node.left,Node.val)
            self.Check(Node.right,Node.val)
            return 1
        else:
            self.Check(Node.left,PrevMaxValue)
            self.Check(Node.right,PrevMaxValue)
            return 0

    #Average faster solution
    good_nodes = 1
    def recursive(self, node, prev_max):
        if not node:
            return
        if prev_max is None:
            prev_max = node.val
        elif prev_max <= node.val:
            self.good_nodes += 1
            prev_max = node.val
        self.recursive(node.left, prev_max=prev_max)
        self.recursive(node.right, prev_max=prev_max)
        
    def goodNodes2(self, root: TreeNode) -> int:
        self.recursive(root, None)
        return self.good_nodes
```

## Path sum III

Given the root of a binary tree and an integer targetSum, return the number of paths where the sum of the values along the path equals targetSum.

The path does not need to start or end at the root or a leaf, but it must go downwards (i.e., traveling only from parent nodes to child nodes).

Examples:

* root = \[10,5,-3,3,2,null,11,3,-2,null,1\], targetSum = 8 -> 3
    * The paths that sum to 8 are shown.
* root = \[5,4,8,11,null,13,4,7,2,null,null,5,1\], targetSum = 22 -> 3

```python
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        self.Counter = 0
        self.targetSum = targetSum

        self.Checks(root,list(),0)
        return self.Counter
    
    def Checks(self,Node,VList,CurrentSum):
        if Node is None:
            return None
        IsLeave = Node.left is None and Node.right is None
        ###
        VList = list(VList)
        NewSum = CurrentSum + Node.val
        VList.append(Node.val)
        if NewSum == self.targetSum:
            self.Counter += 1
        self.ReverseCheck(VList,NewSum)

        if IsLeave:
            return None
        self.Checks(Node.left,VList,NewSum)
        self.Checks(Node.right,VList,NewSum)
        
        return None
    
    def ReverseCheck(self,VList,Total):
        Current = Total
        lasti = len(VList)-1
        for i,n in enumerate(VList):
            Current -= n
            if i == lasti:
                break
            if Current == self.targetSum:
                self.Counter += 1
        return None
    
    def pathSum2(self, root: Optional[TreeNode], targetSum: int) -> int:
        #Most common and faster
        d = {0:1}
        if not root:return 0
        self.f=0
        def dfs(r,s,t):
            if not r:
                return
            else:
                s+=r.val
                self.f+=d.get(s-t,0)
                d[s]=d.get(s,0)+1               
                dfs(r.left,s,t)
                dfs(r.right,s,t)
                d[s]-=1
                s-=r.val
        dfs(root,0,targetSum)
        return self.f
```

## Longest zigzag path in a binary tree

A ZigZag path for a binary tree is defined as follow:

* Choose any node in the binary tree and a direction (right or left).
* If the current direction is right, move to the right child of the current node; otherwise, move to the left child.
* Change the direction from right to left or from left to right.
* Repeat the second and third steps until you can't move in the tree.
* Zigzag length is defined as the number of nodes visited - 1. (A single node has a length of 0).

Return the longest ZigZag path contained in that tree.

Examples:

* root = \[1,null,1,1,1,null,null,1,1,null,1,null,null,null,1\] -> 3
    * Longest ZigZag path in blue nodes (right -> left -> right).
* root = \[1,1,1,null,1,null,null,1,1,null,1\] -> 4
    * Longest ZigZag path in blue nodes (left -> right -> left -> right).
* root = \[1\] -> 0



```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def longestZigZag(self, root: Optional[TreeNode]) -> int:
        #Beats 99.10% at runtime and 93.48% at memory
        if root is None:
            return 0
        self.maxZigZag = 0
        self.Run(root.left,1,'l')
        self.Run(root.right,1,'r')
        return self.maxZigZag
    
    def Run(self,Node,Depth,PreviousDir):
        if Node is None:
            return None
        self.maxZigZag = max(self.maxZigZag,Depth)
        if PreviousDir == 'l':
            self.Run(Node.right,Depth+1,'r')
            self.Run(Node.left,1,'l')
        elif PreviousDir == 'r':
            self.Run(Node.left,Depth+1,'l')
            self.Run(Node.right,1,'r')
```

## Lowest common ancestor of a binary tree

Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

Examples:

* root = \[3,5,1,6,2,0,8,null,null,7,4\], p = 5, q = 1 -> 3
    * The LCA of nodes 5 and 1 is 3.
* root = \[3,5,1,6,2,0,8,null,null,7,4\], p = 5, q = 4 -> 5
    * The LCA of nodes 5 and 4 is 5, since a node can be a descendant of itself according to the LCA definition.
* root = \[1,2\], p = 1, q = 2 -> 1

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        self.pathP = ''
        self.pathQ = ''
        self.P = False
        self.Q = False
        self.p = p
        self.q = q
        self.Run(root,'o')
        CommonPath = self.CommonStrings(self.pathP,self.pathQ)
        return self.followPath(root,CommonPath)
    
    def Run(self,Node,CurrentPath):
        if Node is None:
            return None
        elif Node is self.p:
            self.pathP += CurrentPath
            self.P = True
        elif Node is self.q:
            self.pathQ += CurrentPath
            self.Q = True
        if self.P and self.Q:
            return None
        self.Run(Node.left, CurrentPath + 'l')
        self.Run(Node.right, CurrentPath + 'r')
    
    def CommonStrings(self,S1,S2):
        common = ''
        for i,l1 in enumerate(S1):
            try:
                l2 = S2[i]
            except IndexError:
                break
            if l1 == l2:
                common += l1
            else:
                break
        return common
    
    def followPath(self,root,Path):
        Node = root
        if Path == 'o':
            return Node
        for l in Path:
            if l == 'l':
                Node = Node.left
            elif l == 'r':
                Node = Node.right
        return Node
    
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        #Most common, and better memory and speed efficient
        if not root or root == q or root == p:
            return root
        
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)

        if not left: 
            return right
        if not right:
            return left
        return root
```

Given the root of a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

## Maximum Level Sum of a Binary Tree

Given the root of a binary tree, the level of its root is 1, the level of its children is 2, and so on.

Return the smallest level x such that the sum of all the values of nodes at level x is maximal.

Examples:

* root = \[1,7,0,7,-8,null,null\] -> 2
    * Level 1 sum = 1.
    * Level 2 sum = 7 + 0 = 7.
    * Level 3 sum = 7 + -8 = -1.
    * So we return the level with the maximum sum which is level 2.
* root = \[989,null,10250,98693,-89388,null,null,null,-32127\] -> 2

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxLevelSum(self, root: Optional[TreeNode]) -> int:
        self.maxDepth = 1
        self.maxSum = root.val
        self.BFSLtoR([root])
        return self.maxDepth

    def BFSLtoR(self,CurrentQueue,Depth=1):
        #Beats 84.46 performance and 100% at memory
        if len(CurrentQueue)==0:
            return None
        NewQueue = list()
        currentSum = 0
        for n in CurrentQueue:
            if n is None:
                continue
            currentSum += n.val
            NewQueue.append(n.left)
            NewQueue.append(n.right)
        #print(Depth,currentSum,self.maxSum,self.maxDepth,len(CurrentQueue),len(NewQueue))
        if currentSum > self.maxSum and len(NewQueue) != 0:
            self.maxSum = currentSum
            self.maxDepth = Depth
        self.BFSLtoR(NewQueue,Depth+1)
```
